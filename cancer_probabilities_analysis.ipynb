```python
'''
Cancer data is downloaded from Kaggle (https://www.kaggle.com/datasets/tusharkute/cancer-probabilities)
This is a side-project (data visualisation, ML, etc.)
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

cancer_data = pd.read_csv("cancer_dataset_kaggle\cancer-probabilities.csv")
print(cancer_data)
```

```python
# check attribute type for each features/variables
print(cancer_data.dtypes)
```

Exploratory Analysis: To identify potential correlations between lifestyle habits and cancer risk (https://www.kaggle.com/datasets/tusharkute/cancer-probabilities)

```python
# Investigating potential correlations between categorical lifestyle habits
# and probability of cancer (via boxplots)

# to see correlation between Smoking Habit and cancer probability
sns.boxplot(x="Smoking Habit", y="Probability of Cancer", data=cancer_data)

```

```python
# to see correlation between Drinking Habit and cancer probability
sns.boxplot(x="Drinking Habit", y="Probability of Cancer", data=cancer_data)
```

```python
# to see correlation between Drinking Habit and cancer probability
sns.boxplot(x="Drinking Habit", y="Probability of Cancer", data=cancer_data)
```

```python
# to see correlation between Biking Habit and cancer probability
sns.boxplot(x="Biking Habit", y="Probability of Cancer", data=cancer_data)
```

```python
# to see correlation between Walking Habit and cancer probability
sns.boxplot(x="Walking Habit", y="Probability of Cancer", data=cancer_data)
```

```python
# to see correlation between Jogging Habit and cancer probability
sns.boxplot(x="Jogging Habit", y="Probability of Cancer", data=cancer_data)
```

Finding: 
1. The boxplot between 'Smoking Habit' and 'Probability of Cancer' indicates a positive correlation between the likelihood of getting cancer and the severity of the smoking habit.

2. The boxplot between 'Drinking Habit' and 'Probability of Cancer' also indicates a positive correlation between the likelihood of getting cancer and the frequency of drinking, similar to smoking habit.

3. For biking habit, we can see a negative correlation between the frequency of biking habit and the likelihood of developing cancer. This can be seen as people with high biking frequency has the least likelihood of getting cancer, and vice versa. Unlike correlations between smoking habit and drinking habit with regards to probability of cancer.

4. There is also a negative correlation seen between walking habit and probability of cancer, as people who frequently walk have the least likelihood of developing cancer. But people who rarely walk have higher likelihood of developing cancer.

5. There is also a negative correlation seen between jogging habit and probability of cancer, as the frequency of jogging increases, the likelihood of developing cancer decreases, and vice versa.
